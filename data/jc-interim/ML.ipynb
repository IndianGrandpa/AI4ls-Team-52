<h1>Load 2018 data </h1>

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_2009_path = "..\df_2009.csv"
df_2015_path = "..\df_2015.csv"
df_2018_path = "..\df_2018_with_bd_erosion.csv"
metadata_path = "sra_metadata_5d8517e0-8bb8-4a34-ba02-26c397dea5a6.csv"
```

```python
df_2009 = pd.read_csv(df_2009_path, delimiter=",")
df_2015 = pd.read_csv(df_2015_path, delimiter=",")
df_2018 = pd.read_csv(df_2018_path, delimiter=",")
df_metadata = pd.read_csv(metadata_path, delimiter=",")
```

```python
df_soil = pd.merge(df_2015, df_2018, how='outer', on='point_id')
df_soil_total = pd.merge(df_soil, df_2009, how='outer', on='point_id')
df_soil_total
```

```python
df_soil_ml = df_soil_total.loc[df_soil_total['ph_2009'].notnull() & df_soil_total['ph_2015'].notnull() & df_soil_total['ph_2018'].notnull()]
df_soil_ml
```

```python
df_soil_ml.to_csv("merged_soil_intersects.csv", index=False)
```

```python
df_soil_ml.isna().sum()
```

```python
df_soil_ml.loc[:, 'survey_erosion_signs'] = df_soil_ml['survey_erosion_signs'].fillna(0)

mean_bd_10 = df_soil_ml['bd_0_10'].mean()
mean_bd_20 = df_soil_ml['bd_10_20'].mean()

df_soil_ml.loc[:, ['bd_0_10']] = df_soil_ml['bd_0_10'].fillna(mean_bd_10)
df_soil_ml.loc[:, ['bd_10_20']] = df_soil_ml['bd_10_20'].fillna(mean_bd_20)

```

```python
# LODs
df_soil_ml['caco3_2018'].replace("<  LOD", 0.1, inplace=True)
mean_caco3 = df_soil_ml['caco3_2018'].notnull().mean()
df_soil_ml['caco3_2018'].fillna(mean_caco3, inplace=True)

mean_ec = df_soil_ml['ec_2018'].mean()
df_soil_ml['ec_2018'].fillna(mean_ec, inplace=True)

df_soil_ml['p_2018'].replace("< LOD", 0, inplace=True)
mean_p = df_soil_ml['p_2018'].notnull().mean()
df_soil_ml['p_2018'].fillna(mean_p, inplace=True)
```

```python
df_soil_ml.to_csv("merged_soil_intersects_nonan.csv", index=False)
```
