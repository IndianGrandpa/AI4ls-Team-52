```python
def satrification(df,column):
    value_list=df[column].value_counts().sort_values(ascending=False)
    df_0=df[df[column]==value_list.index[0]]
    df_1=df[df[column]==value_list.index[1]]
    df0_sampled = df_0.sample(n=len(df_1), random_state=2).copy()
    df_new=pd.concat([df_1,df0_sampled],axis=0)
    return df_new
df_ml=satrification(df,"survey_erosion_signs")
```
